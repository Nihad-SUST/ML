```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
X = np.array([[1000], [1200], [1500], [1800], [2000], [2200]])  # House size (independent variable)
y = np.array([150, 180, 210, 250, 280, 300])                # House price (dependent variable)

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict for a new house size (1600 sq. ft.)
predicted_price = model.predict([[2000]])
print(f"Predicted price for 2100 sq. ft. house: ${predicted_price[0] * 1000:.2f}")

# Visualization
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.scatter(2000, predicted_price, color='green', label='Prediction (1600 sq. ft.)')
plt.xlabel('House Size (sq. ft.)')
plt.ylabel('House Price ($1000s)')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Dataset (independent variables X and target y)
X = np.array([
    [1000, 2, 20],
    [1500, 3, 15],
    [1800, 3, 10],
    [2000, 4, 5],
    [2300, 4, 2]
])

y = np.array([150, 200, 250, 300, 350])  # House prices in $1000s

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict price for a 1600 sq. ft. house with 3 bedrooms and 10 years old
new_house = np.array([[1600, 3, 10]])
predicted_price = model.predict(new_house)

print(f"Predicted price for 1600 sq. ft., 3-bed, 10-year-old house: ${predicted_price[0] * 1000:.2f}")

# Model coefficients and intercept
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficients: {model.coef_}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([[2], [3], [5], [7], [9]])  # Study hours
y = np.array([50, 60, 75, 85, 95])       # Exam scores

# Model
model = LinearRegression()
model.fit(X, y)

# Predict score for 4 hours of study
predicted_score = model.predict([[4]])
print(f"Predicted score for 4 hours of study: {predicted_score[0]:.2f}%")

# Plot
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.scatter(4, predicted_score, color='green', label='Prediction (4 hours)')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score (%)')
plt.legend()
plt.show()
```

```python
X = np.array([[1], [3], [5], [7], [10]])  # Years of experience
y = np.array([30, 50, 70, 90, 120])       # Salary in $1000s

# Model
model = LinearRegression()
model.fit(X, y)

# Predict salary for 6 years of experience
predicted_salary = model.predict([[6]])
print(f"Predicted salary for 6 years of experience: ${predicted_salary[0] * 1000:.2f}")

# Plot
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.scatter(6, predicted_salary, color='green', label='Prediction (6 years)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary ($1000s)')
plt.legend()
plt.show()
```

```python
#Python Code for Multiple Linear Regression:
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Dataset (independent variables X and target y)
X = np.array([
    [1000, 2, 20],
    [1500, 3, 15],
    [1800, 3, 10],
    [2000, 4, 5],
    [2300, 4, 2]
])

y = np.array([150, 200, 250, 300, 350])  # House prices in $1000s

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict price for a 1600 sq. ft. house with 3 bedrooms and 10 years old
new_house = np.array([[1600, 3, 10]])
predicted_price = model.predict(new_house)

print(f"Predicted price for 1600 sq. ft., 3-bed, 10-year-old house: ${predicted_price[0] * 1000:.2f}")

# Model coefficients and intercept
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficients: {model.coef_}")
```

```python
#Scikit-Learn Code for Both Examples
#Here's the Python code to verify the calculations using scikit-learn:

import numpy as np
from sklearn.linear_model import LinearRegression

# Example 1: Study Hours vs. Exam Score
X1 = np.array([[2], [3], [5], [7], [9]])
y1 = np.array([50, 60, 75, 85, 95])

model1 = LinearRegression()
model1.fit(X1, y1)

print(f"Example 1: Study Hours vs. Score")
print(f"Slope (b1): {model1.coef_[0]:.3f}")
print(f"Intercept (b0): {model1.intercept_:.3f}")
print(f"Predicted score for 4 hours of study: {model1.predict([[4]])[0]:.2f}%\n")

# Example 2: Experience vs. Salary
X2 = np.array([[1], [3], [5], [7], [10]])
y2 = np.array([30, 50, 70, 90, 120])

model2 = LinearRegression()
model2.fit(X2, y2)

print(f"Example 2: Experience vs. Salary")
print(f"Slope (b1): {model2.coef_[0]:.3f}")
print(f"Intercept (b0): {model2.intercept_:.3f}")
print(f"Predicted salary for 6 years of experience: ${model2.predict([[6]])[0] * 1000:.2f}")
```
